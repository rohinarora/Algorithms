```python
class Solution:
    # @param A : string
    # @return an integer
    def power(self, A):
        A=int(A)
        if A==1:
            return 0
        while (A>1):
            rem=A%2
            A=A/2
            #print (rem,A)
            if (rem%2)!=0:
                return 0
        return 1
```
